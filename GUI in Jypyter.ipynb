```python
import ipywidgets as wid
from ipywidgets import interactive

name_input = None
age_input = None

def set_name(name):
    return name

def set_age(age):
    return age

def print_person(b):
    name = name_input.children[0].value
    age = age_input.children[0].value
    print(f"Your name is {name_input.result}, and your age is {age_input.result}")


"""wid.Text(
    value= "",
    placeholder="Enter Name",
    description="Name",
    disabled=False

)"""
name_input = wid.interactive(set_name, name=wid.Text(value= "",
    placeholder="Enter Name",
    description="Name",
    disabled=False))
display(name_input)

age_input = wid.interactive(set_age, age=wid.IntSlider(
    value=7,
    min=18,
    max=90,
    step=1,
    description='Age:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'))
display(age_input)


button = wid.Button(
    description="Continue",
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me',
    icon='check'
                    )
button.on_click(print_person)
display(button)
```
